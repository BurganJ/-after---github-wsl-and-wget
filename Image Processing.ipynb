```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import cmocean 
import glob
import pandas as pd
```

```python
fns = glob.glob(r"C:\Users\azale\algorithms\Images\*.png")
df = pd.DataFrame()
dict_keys = ['filename','max','median','min']
for fn in fns:
    I= imageio.v3.imread(fn)
    
    dbz = (I.astype(float)/2 - 33)
    dbz[dbz < -32] = np.nan
    max = np.nanmax(dbz)
    min = np.nanmin(dbz)
    med = np.nanmedian(dbz)

    dict_values = [fn,max,med,min]
    zip_object=zip(dict_keys,dict_values)
    dict_object= dict(zip_object)
    df = df.append(dict_object,ignore_index = True)
    
    
```
