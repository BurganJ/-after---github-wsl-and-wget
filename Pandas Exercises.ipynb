```python
#(1) Import numpy as np, matplotlib.pyplot as plt, and pandas as pd.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#(2) Read the Fisher Iris Data in as a variable called df.
df = pd.read_csv("iris.csv")
```

```python
#(3) Display the first five lines of df using head.
print(df.head(5))
```

```python
#(4-7) - Plot a histogram of each of the four variables (leaf/sepal length/width). Adjust the number of bins to produce a resolution you find instructive/visually satisfying. Use xlabel and ylabel to label the axes. Each plot should be its own codeblock.
hist1 = plt.hist(df.petal_width, bins = 20, ec = 'black')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
```

```python
#(5)
hist2 = plt.hist(df.petal_length, bins = 20, ec = 'black')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
```

```python
#(6)
hist3 = plt.hist(df.sepal_width, bins = 20, ec = 'black')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
```

```python
#(7)
hist4 = plt.hist(df.sepal_length, bins = 20, ec = 'black')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
```

```python
#(8) Save each figure using savefig.
plt.savefig('hist1')
plt.savefig('hist2')
plt.savefig('hist3')
plt.savefig('hist4')
```

```python
#(9) Using groupby, group the data by species
grp = df.groupby(by = 'species')
```

```python
#(10) Using the groups, create a bar graph of the mean sepal length by species. Label the axes.
grp.mean().plot(kind = 'bar', y = 'sepal_length', ylabel = 'Length', xlabel = 'Species')
```

```python
#(11) Use subplots to make a 2x2 pane figure, with each pane a scatterplot of your choosing. Make sure axes are labeled!
plt.subplot(2,2,1)
plt.scatter(df.petal_width, df.petal_length)
plt.xlabel('Petal Width')
plt.ylabel('Petal Length')

plt.subplot(2,2,2)
plt.scatter(df.sepal_width, df.sepal_length)
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')

plt.subplot(2,2,3)
plt.scatter(df.petal_width, df.sepal_width)
plt.xlabel('Petal Width')
plt.ylabel('Sepal Width')

plt.subplot(2,2,4)
plt.scatter(df.petal_length, df.sepal_length)
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
```

```python
#(12) Use tail to print out the last 10 rows of df.
print(df.tail(10))
```

```python
#(13) Use a combination of np.sum() and a selector (e.g., (df['x'] > 3), count how many rows have a petal length greater than 40.
arr = np.array(df)
pl_40 = np.sum(arr[:,2] > 40)

print(pl_40)
```

```python
#(14) How many have a sepal width below the median sepal width?
median_sw = np.median(arr[:,3])
sw_below_med = np.sum(arr[:,3] < median_sw)
print(sw_below_med)
```

```python
#(15) Use groups to report the mean values for each variable (by species).
print(df.groupby(by = 'species').mean())
```

```python
#(16) Calculate a new variable based on existing variables, and create a histogram of its distribution.
#difference between sepal length and petal length
diff = df.sepal_length - df.petal_length
#seeing if any values are negative
def is_negative():
    if any in diff < 0:
        return True
    else:
        return False
print(is_negative())

plt.hist(diff, bins = 20, ec = 'black')
plt.xlabel('Difference Between Sepal Length and Petal Length')
plt.ylabel('Frequency')
```

```python
#(17) Export the data with the new calculated variable as an Excel file. Open it in Excel, and comment on its appearance. Look at the documentation of to_excel and figure out how to make it so the index values (line numbers) don't get exported.
df['sl - pl'] = diff
print(df)
df.to_csv('Iris_New.csv')
#now has column of index values
#remove index values
df.to_excel('Iris_New.xlsx', index = False)
```
